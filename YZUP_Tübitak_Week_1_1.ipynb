# SQL JSON PARQUET

sqlite3 sqlalchemy duckdb mangodb gibi pythonda ilişkisel olan ve olmayan veri tabanları için paketler mevcut.

```python
import os
from sqlalchemy import create_engine

DataFolderName="../data-type"
```

```python
def rfd(filename):
  return os.path.join(DataFolderName,filename)  # Sistemin bağımsız olması için os.path.join kullanıldı,os farketmeksizin kodun çalışmasını sağlar.

if not os.path.exists(DataFolderName):
  os.mkdir(DataFolderName)

os.listdir(DataFolderName)
```

```python
kismi_xls.to_csv(rfd("csv-example.csv"))
kismi_xls.to_json(rfd("json-example.json"))
kismi_xls.to_parquet(rfd("parquet-example.parquet"),engine='auto',index=False)  #Parquet veriler daha az yer kaplar.
```

```python
engine=create_engine(os.path.join("sqlite:///",DataFolderName,"sql-example.sqlite"),echo=True)
```

```python
kismi_xls.to_sql("kismi-veri",con=engine,index=False)
```

```python
#Birden fazla eklenebilir.
engine=create_engine(os.path.join("sqlite:///",DataFolderName,"sql-example-multi-sqlite.sqlite"),echo=True)
kismi_xls.to_sql("kismi-veri",con=engine,index=False)
xls.to_sql("Full veri",con=engine,index=False)
```

```python
engine=create_engine(os.path.join("sqlite:///",DataFolderName,"sql-example.sqlite"),echo=False)
```

```python
data_sql=pd.read_sql("kismi-veri",con=engine)
```
