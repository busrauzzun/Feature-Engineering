```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('boston.csv')
data.describe()
```

```python
data.head()
```

```python
x=data.drop('MEDV',axis=1)
y=data[['MEDV']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
X_train[['LSTAT','NOX','DIS']]
```

### Eşit Aralıklarla Ayrıklaştırma

```python
from sklearn.preprocessing import KBinsDiscretizer

dsc=KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='uniform')
X_train[['LSTAT','NOX','DIS']]=dsc.fit_transform(X_train[['LSTAT','NOX','DIS']])
X_test[['LSTAT','NOX','DIS']]=dsc.transform(X_test[['LSTAT','NOX','DIS']])
```

```python
X_train['LSTAT'].value_counts()
```

```python
dsc.bin_edges_
```

### Eşit Ağırlıklı Ayrıştırma

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)

dsc=KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='quantile')
X_train[['LSTAT','NOX','DIS']]=dsc.fit_transform(X_train[['LSTAT','NOX','DIS']])
X_test[['LSTAT','NOX','DIS']]=dsc.transform(X_test[['LSTAT','NOX','DIS']])
```

```python
X_train[['LSTAT','NOX','DIS']]
X_test[['LSTAT','NOX','DIS']]
```

```python
X_train['LSTAT'].value_counts()
```

### K-Means ile Ayrıklaştırma

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)

dsc=KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='kmeans')
X_train[['LSTAT','NOX','DIS']]=dsc.fit_transform(X_train[['LSTAT','NOX','DIS']])
X_test[['LSTAT','NOX','DIS']]=dsc.transform(X_test[['LSTAT','NOX','DIS']])
```

```python
X_train[['LSTAT','NOX','DIS']]
X_test[['LSTAT','NOX','DIS']]
```

```python
X_train['LSTAT'].value_counts()
```

### Karar Agaclari ile Ayriklaştırma

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

from sklearn.tree import DecisionTreeRegressor


discretizer= DecisionTreeRegressor(max_depth=5, random_state=0)

discretizer.fit(X_train[['LSTAT']], y_train)
X_train['LSTAT_tree']=discretizer.predict(X_train[['LSTAT']])
X_test['LSTAT_tree']=discretizer.predict(X_test[['LSTAT']])
```
